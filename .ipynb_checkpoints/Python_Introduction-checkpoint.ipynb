# Section 5: Basics of Python 

Let's move on to writing some Python code. Python is a high-level programming language that is known for its simplicity and readability. Its a versatile language that is used in a wide range of applications from web development to data analysis.

In this section, we will look at a few basic concepts in Python, such as variables, data types, and operations. These are the building blocks that will allow you to write more complex programs later on.

First, let's talk about variables. In Python, a variable is a name that is used to denote something or a value. You can think of it as a tag that you attach to a value.

Now, let's move on to data types. In Python, every value has a data type. There are several basic data types in Python, including integers, floating point numbers, strings, and Boolean values.

Let's also discuss operations. Python provides a variety of operations that can be performed on values, including mathematical operations, comparison operations, and logical operations.

Now, let's see some examples of these concepts in action.


````python
# Variables
x = 5
y = "Hello, World!"

# Print the variables
print(x)
print(y)

# Data Types
print(type(x))  # Integer
print(type(y))  # String

# Operations
z = x + 10
print(z)  # Mathematical operation

a = x > 3
print(a)  # Comparison operation

b = x > 3 and x < 10
print(b)  # Logical operation
````

## Section 6: Lists in Python

Python lists are used to store multiple items in a single variable. They are one of four built-in data types in Python used to store collections of data. The other three are Tuple, Set, and Dictionary, all with different qualities and usage.

Lists are created by placing all the items (elements) inside square brackets `[]`, separated by commas.

In the upcoming cells, we will learn how to create a list, access its elements, modify them, and add new elements.


````python
# Creating a list
fruits = ["apple", "banana", "cherry"]
fruits
````

````python
# Accessing elements in a list
print("First fruit: ", fruits[0])
print("Second fruit: ", fruits[1])
print("Third fruit: ", fruits[2])
````

````python
# Modifying elements in a list
fruits[1] = "blueberry"
fruits
````

````python
# Adding elements to a list
fruits.append("dragonfruit")
fruits
````

## Section 7: Conditional Statements in Python

Conditional statements in Python allow us to test conditions and change the behavior of the program accordingly. These include `if`, `elif`, and `else` statements. 

- The `if` statement is used to test a specific condition. If the condition is true, the block of code under the `if` statement will execute.
- The `elif` statement allows us to check multiple expressions for TRUE and execute the first one that is true.
- The `else` statement catches anything which isn't caught by the preceding conditions.

Let's look at some examples of conditional statements.


````python
# Defining a variable
x = -10

# Using if statement
if x > 0:
    print("x is positive")
````

````python
# Using if-else statement
if x % 2 == 0:
    print("x is even")
else:
    print("x is odd")
````

````python
# Using if-elif-else statement
if x < 0:
    print("x is negative")
elif x > 0:
    print("x is positive")
else:
    print("x is zero")
````

## Section 8: Loops in Python

Loops are a way to repeatedly execute some code. Here's the syntax for `for` and `while` loops in Python:

- `for` loops are traditionally used when you have a block of code which you want to repeat a fixed number of times.

```python
for variable in sequence:
    # code to be executed


- `while` loops are used when you want to repeat a block of code an unknown number of times, as long as some condition is met.

```python
while condition:
    # code to be executed


````python
# Defining a list
numbers_list = [1, 2, 3, 4, 5]

# Using for loop to iterate over the list
for number in numbers_list:
    print("number =", number)
````

````python
# Using while loop
i = 0
while i < 5:
    print("i:",i)
    i = i + 1
````

# Basic Plotting with Matplotlib

Matplotlib is a plotting library for Python. It is used for creating static, animated, and interactive visualizations in Python. 

`pyplot` is a module in Matplotlib which provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python, with the advantage that it is free, open-source, and most importantly, a more powerful programming language.

## Creating a Simple Plot

To create a simple plot using Matplotlib, you need to import the `pyplot` module from the Matplotlib library. Here's an example:



````python
# Import the required module for plotting
import matplotlib.pyplot as plt

# Data for the x-axis
x = [1, 2, 3, 4, 5]

# Data for the y-axis
y = [2, 4, 6, 8, 10]

# Create a simple line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
````

This code creates a simple line plot using Matplotlib. We define two lists, x and y, to represent the x-axis and y-axis values. The `plot()` function is used to create the line plot by passing these lists as arguments. Then, we add labels to the x-axis and y-axis using the `xlabel()` and `ylabel()` functions, and set a title for the plot using the `title()` function. Finally, the `show()` function is used to display the plot.
